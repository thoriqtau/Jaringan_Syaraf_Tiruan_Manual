# STUDI KASUS LOGIKA NAND DENGAN EPOCH 100

```python
# fungsi sigmoid
def sigmoid(x):
    return 1 / (1 + 2.71828 ** (-x))

# input ke hidden layer
w11,w12,w21,w22 = 0, 0, 0, 0

# bias ke hidden layer
w31, w32 = 0, 0

# hidden layer ke output
v11, v12 = 0,0

# bias ke output
v21 = 0

# data input
x1 = [0,1,0,1]
x2 = [0,0,1,1]

# data bias
bias = 1

# data target
target = [1,1,1,0]

# learning rate
learning_rate = 1

# epoch
epoch = 100

for i in range(epoch):
  total_error = 0
  for j in range(len(target)):
    # feedforward
    Z_1 = (w31 + (((x1[j]) * w11) + ((x2[j]) * w21)))
    Z_1 = sigmoid(Z_1)

    Z_2 = (w32 + (((x1[j]) * w12) + ((x2[j]) * w22)))
    Z_2 = sigmoid(Z_2)

    y = (v13 + (((Z_1) * v11) + ((Z_2) * v12)))
    y = sigmoid(y)

    # feedback (backpropagation)
    error = 0.5 * (target[j] - y) ** 2
    total_error += error

    # update bobot (hidden layer ke output layer)
    sigmoid_1 = (target[j]-y)*y*(1-y)

    delta_v11 = learning_rate * sigmoid_1 * Z_1
    delta_v12 = learning_rate * sigmoid_1 * Z_1
    delta_v13 = learning_rate * sigmoid_1 * bias

    # update bobot (input layer ke hidden Layer)
    sigmoid_net_1 = sigmoid_1 * v11
    sigmoid_net_2 = sigmoid_1 * v12

    s1 = sigmoid_net_1 * Z_1 * (1-Z_1)
    s2 = sigmoid_net_2 * Z_2 * (1-Z_2)

    delta_w11 = learning_rate * s1 * x1[j]
    delta_w12 = learning_rate * s2 * x2[j]

    delta_w21 = learning_rate * s1 * x1[j]
    delta_w22 = learning_rate * s2 * x2[j]

    delta_w31 = learning_rate * s1 * bias
    delta_w32 = learning_rate * s2 * bias

    # Update bobot
    v11 = v11 + delta_v11
    v12 = v12 + delta_v12
    v13 = v13 + delta_v13

    w11 = w11 + delta_w11
    w12 = w12 + delta_w12

    w21 = w21 + delta_w21
    w22 = w22 + delta_w22

    w31 = w31 + delta_w31
    w32 = w32 + delta_w32

    # hasil prediksi data ke-
    print(f"Input: {x1[j]},{x2[j]} - Prediksi: {round(y)} - Jawaban: {target[j]}, Error: {error:.2f}")
  # epoch ke-
  print(f"Epoch {i+1}\n")
```
